```python
# Sample data (replace this with your actual dataset)
data = {
    'Gender': ['Male', 'Female', 'Male', 'Male', 'Female', 'Female', 'Male', 'Male', 'Female', 'Male' ]
}

#Create DataFrame
df = pd.DataFrame(data)

#Count occurences of each gender
gender_counts = df['Gender'].value_counts()

# Plot bar graph
plt.figure(figsize=(8, 6))
gender_counts.plot(kind='bar', color='skyblue')

#Add labels and title
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Distribution of Genders')

#Show plot
plt.show()
```

```python
# Sample data (replace this with your actual dataset)
data = {
    'Age': [45, 50, 35, 60, 55, 40, 30, 65, 70, 25],
    'Total Bilirubin': [1.2, 2.5, 3.0, 1.8, 2.0, 1.5, 2.2, 2.8, 3.5, 1.0]
}

#Create DataFrame
df = pd.DataFrame(data)

#Extract variable for scatter plot
x = df['Age']
y = df['Total Bilirubin']

#Create scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='orange', alpha=0.7)

# Add labels and title
plt.xlabel('Age')
plt.ylabel('Total Bilrubin')
plt.title('Scatter plot: Age vs Total Bilrubin')

#Show plot
```

```python


# Sample data (replace this with your actual dataset)
data = {
    'Gender': ['Male', 'Female', 'Male', 'Male', 'Female', 'Female', 'Male', 'Male', 'Female', 'Male'],
    'SGPT': [35, 45, 40, 50, 55, 42, 48, 38, 60, 30]
}

# Create DataFrame
df = pd.DataFrame(data)

# Create violin plot
plt.figure(figsize=(10, 6))
sns.violinplot(x='Gender', y='SGPT', data=df, palette='muted')

# Add labels and title
plt.xlabel('Gender')
plt.ylabel('SGPT')
plt.title('Violin plot: SGPT Distribution by Gender')

# Show plot
plt.show()
```
